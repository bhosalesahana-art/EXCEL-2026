```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data={'StudyHours':[2,4,6,8,10],
      'Marks':[35,50,65,80,90],
      'SleepHours':[8,7,6,5,4]}
d=pd.DataFrame(data)
print('Dataset:\n',d)
print("\nCorrelation Matrix:\n",d.corr())
cor_value=d['StudyHours'].corr(d['Marks'])
print('Correlation Between StudyHours and Marks\n',cor_value)
print('\n Numpy Correlation:\n',np.corrcoef(d['StudyHours'],d['Marks']))
plt.figure()
plt.scatter(d['StudyHours'],d['Marks'])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("StudyHours Vs Marks")
plt.show()
plt.figure()
sns.heatmap(d.corr(),annot=True)
plt.title("Correlation Heatmap")
plt.show()
print("\n Pearson Correlation:\n",d.corr(method='pearson'))
print("\nSpearman Correlation:\n",d.corr(method='spearman'))
print("\nKendall Correlation:\n",d.corr(method='kendall'))

```
