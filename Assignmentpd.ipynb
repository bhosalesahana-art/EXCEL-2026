```python
import pandas as pd
```

```python
data=pd.read_csv('sample_data/Book2.csv')
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.fillna('Not Avaliable')
```

```python
data['Salary']=data['Salary'].fillna(data['Salary'].mean())
data
```

```python
data['Department']=data['Department'].fillna(data['Department'].mode()[0])
data
```

```python
data['Experience']=data['Experience'].fillna(data['Experience'].median())
data
```

```python
data=data.dropna(subset=['Name'])
data
```

```python
data.duplicated()
```

```python
data['Emp_Id'].duplicated()
```

```python
data.drop_duplicates()
```

```python
data=data.drop_duplicates(subset='Emp_Id',keep='first')
data
```

```python
data=data.reset_index(drop=True)
data
```
