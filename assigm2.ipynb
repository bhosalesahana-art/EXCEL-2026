```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('sample_data/netflix_titles.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data=data.dropna(subset=['release_year','rating','country'])
data
```

```python
type1=data['type'].value_counts()
type1.plot(kind='bar',title='Movies VS TVshows on Netflix')
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
year=data['release_year'].value_counts().sort_index()
year.tail(15).plot(kind='line',marker='o',title='Netflix Content Released in Recent Years')
plt.xlabel('Release year')
plt.ylabel('Number of titles')
plt.show()
```

```python
rating=data['rating'].value_counts().head(10)
rating.plot(kind='pie',autopct='%1.1f%%',title='Top 10 Content Ratings')
plt.ylabel("")
plt.show()
```

```python
country=data['country'].value_counts().head(5)
country.plot(kind='bar',title='Top 5 Content Producing Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.show()
```
