```python
import pandas as pd
```

```python
data=[10,12,11,13,12,14,100]
d1=pd.DataFrame(data,columns=['Marks'])
print(d1)
```

```python
Q1=d1['Marks'].quantile(0.25)
Q3=d1['Marks'].quantile(0.75)
print('Q1:',Q1)
print('Q3:',Q3)
```

```python
IQR=Q3-Q1
print('IQR:',IQR)
```

```python
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print('lower bound:',lower_bound)
print('upper bound:',upper_bound)
```

```python
outliers=d1[(d1['Marks']<lower_bound)|(d1['Marks']>upper_bound)]
print("Outliers:")
print(outliers)
```

```python
from scipy import stats
import pandas as pd
data=pd.DataFrame([10,12,11,13,12,14,100],columns=['Marks'])
data['z_score']=stats.zscore(data['Marks'])
print(data)
```

```python
outliers=data[data['z_score'].abs()>2]
print(outliers)
```
