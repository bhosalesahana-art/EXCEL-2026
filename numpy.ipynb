```python
import numpy as np
print("=====STUDENT ANALYSIS USING NUMPY====\n")
#1.Dataset Creation(with missing values)
marks=np.array([[78,85,90],[88,76,np.nan],[67,70,80],[90,95,88],[np.nan,60,75]])
subjects=["Math","Science","English"]
print("Original Marks Data:\n",marks,"\n")
```

```python
# 2 Array Properties
print("Array Shape:",marks.shape)
print("Array Dimensions:",marks.ndim)
print("Total Elements:",marks.size)
print("Data type:",marks.dtype,"\n")
```

```python
#3 Handling Missing values
subject_mean=np.nanmean(marks,axis=0)
missing_indices=np.where(np.isnan(marks))
marks[missing_indices]=np.take(subject_mean,missing_indices[1])
print("Marks after handling missing values:\n",marks,"\n")
```

```python
# 4 indexing & Slicing
print("Math marks:",marks[:,0])
print("Marks of student2:",marks[1],"\n")
```

```python
# 5 BoardCasting(Adding Bonus Marks)
bonus_marks=marks+5
print("Marks after adding bouns:\n",bonus_marks,"\n")
```

```python
# 6 Statistical Analysis
student_avg=marks.mean(axis=1)
subject_avg=marks.mean(axis=0)
print("Student-wise Average Marks:",student_avg)
print("Subject-wise Average Marks:")
for i in range(len(subjects)):
  print(subjects[i],":",subject_avg[i])
print()
```

```python
# 7 Performance Metrics
print("Highest Score:",marks.max())
print("Lowest Score:",marks.min())
print("Standard Deviation:",marks.std(),"\n")
```

```python
# 8 Ranking Students
ranking=np.argsort(student_avg)[::-1]
print("Student Ranking(Top to Bottom):")
for i,rank in enumerate(ranking):
  print(f"Rank{i+1}->Student{rank+1}")
print()
```

```python
# 9 Boolean Indexing(top performers)
top_students=student_avg>85
print("Top Performing Students(Avg>85):",top_students)
print("Student Nummbers:",np.where(top_students)[0]+1,"\n")
```

```python
# 10 Shape Manipulation
flat_marks=marks.flatten()
print("Flattened Marks Data:",flat_marks,"\n")
```

```python
#11 Sorting &Searching
sorted_marks=np.sort(flat_marks)
print("Sorted Marks:",sorted_marks)
print("Unique Marks:",np.unique(sorted_marks),"\n")
```

```python
#12 Random Data Simulation
np.random.seed(10)
random_scores=np.random.randint(60,100,size=(5,3))
print("Randomly Generated Scores:\n",random_scores,"\n")
```

```python
#13 Linear Algebra(Total Score Calclution)
weights=np.array([1,1,1])
total_scores=np.dot(marks,weights)
print("Total Scores of Students:",total_scores,"\n")
```

```python
#14 Performance Comparison(list vs Numpy)
python_list=[1,2,3,4,5]
numpy_array=np.array(python_list)
print("Python List*2:",[x*2 for x in python_list])
print("Numpy Array*2:",numpy_array*2)
```
